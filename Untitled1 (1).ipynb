```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import mutual_info_classif
from skfeature.function.similarity_based import fisher_score
import matplotlib.pyplot as plt
from sklearn.feature_selection import chi2
from sklearn.linear_model import LogisticRegression, LinearRegression
```

```python
df = pd.read_excel('GermanCreditData.xlsx', engine='openpyxl')
df['Purpose'] = df['Purpose'].replace(['X'], 10)


X_coulumns = ['Duration', 'History', 'Purpose', 'Amount', 'Savings',
              'Emploed', 'Installp', 'marital', 'Coapp', 'Resident', 'Property',
              'Age', 'Other', 'housing', 'Existcr', 'Job', 'Depends', 'Telephone',
              'Foreign', 'Bad', 'Good']
Y_coulumns = ['Checking']
continuous_var = ['Amount', 'Age']
```

```python
df1 = df.loc[df['Checking'].isin([1, 2])]
df1_x = df1[X_coulumns]
df1_y = df1[Y_coulumns]

df2 = df.loc[df['Checking'].isin([3, 4])]
df2_x = df2[X_coulumns]
df2_y = df2[Y_coulumns]
```

```python
df1_X_train, df1_X_test, df1_y_train, df1_y_test = train_test_split(df1_x, df1_y, test_size=0.3, random_state=0, stratify=df1_y)
df2_X_train, df2_X_test, df2_y_train, df2_y_test = train_test_split(df2_x, df2_y, test_size=0.3, random_state=0, stratify=df2_y)
```

```python
chi_scores = chi2(df1_X_train,df1_y_train)
p_values = pd.Series(chi_scores[1],index = X_coulumns)
p_values.sort_values(ascending = False , inplace = True)
p_values.plot.bar()
```

```python
chi_scores = chi2(df2_X_train,df2_y_train)
p_values = pd.Series(chi_scores[1],index = X_coulumns)
p_values.sort_values(ascending = False , inplace = True)
p_values.plot.bar()
```

```python
df1_features = ["Amount", "Purpose", "Savings", "Resident"]
df2_features = ["Amount", "Duration", "Savings", "Bad"]
```

```python
logisticRegr = LogisticRegression()
logisticRegr.fit(df1_X_train[df1_features], df1_y_train)
predictions_df1 = logisticRegr.predict(df1_X_test[df1_features])

score = logisticRegr.score(df1_X_test[df1_features], df1_y_test)
print(score)
```

```python
logisticRegr = LogisticRegression()
logisticRegr.fit(df2_X_train[df2_features], df2_y_train)

score = logisticRegr.score(df2_X_test[df2_features], df2_y_test)
print(score)
```

```python
linearRegr = LinearRegression()
linearRegr.fit(df1_X_train[df1_features], df1_y_train)

score = linearRegr.score(df1_X_test[df1_features], df1_y_test)
print(score)
```
