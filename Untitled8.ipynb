```python
Q1=dfcheck12['Age'].quantile(0.25)
Q3=dfcheck12['Age'].quantile(0.75)
IQR=Q3-Q1
IQR
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 +(whisker_width*IQR)
dfcheck12['Age']=np.where(dfcheck12['Age']>upper_whisker,upper_whisker,np.where(dfcheck12['Age']<lower_whisker,lower_whisker,dfcheck12['Age']))
```

```python
# Print outlier range for different variables and remove them from Exel directly (<= 12 months)

# Age
Age_IQR = dfcheck12.describe()['Age'][6]-dfcheck12.describe()['Age'][4]
Age_L = dfcheck12.describe()['Age'][4]-1.5*Age_IQR
Age_H = dfcheck12.describe()['Age'][6]+1.5*Age_IQR
print('Check=1,2 Age:', Age_L, '//', Age_H)

# Amount
Amount_IQR = dfcheck12.describe()['Amount'][6]-dfcheck12.describe()['Amount'][4]
Amount_L = dfcheck12.describe()['Amount'][4]-1.5*Amount_IQR
Amount_H = dfcheck12.describe()['Amount'][6]+1.5*Amount_IQR
print('Check=1,2 Amount:', Amount_L, '//', Amount_H)

# Existcr
Existcr_IQR = dfcheck12.describe()['Existcr'][6]-dfcheck12.describe()['Existcr'][4]
Existcr_L = dfcheck12.describe()['Existcr'][4]-1.5*Existcr_IQR
Existcr_H = dfcheck12.describe()['Existcr'][6]+1.5*Existcr_IQR
print('Check=1,2 Existcr:', Existcr_L, '//', Existcr_H)

# Duration
Duration_IQR = dfcheck12.describe()['Duration'][6]-dfcheck12.describe()['Duration'][4]
Duration_L = dfcheck12.describe()['Duration'][4]-1.5*Duration_IQR
Duration_H = dfcheck12.describe()['Duration'][6]+1.5*Duration_IQR
print('Check=1,2 Duration:', Duration_L, '//', Duration_H)
```
